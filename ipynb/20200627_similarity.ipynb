```python
import pandas as pd
```

```python
similarity = pd.read_feather('../data/similarity')

similarity.head(10)
```

```python
tweets = pd.read_feather('../data/tweets_limpios')
tweets = tweets[tweets['text'] != '']
tweets
```

```python
tweets_ods = pd.DataFrame(tweets['text'], index=tweets['index'])
tweets_ods.head()
```

```python
minim=0.7

def comprovarValor(i):
  if(i>=minim):
    return 1
  return 0

for i in similarity.columns:
  tweets_ods['ODS_'+i] = similarity[i].apply(comprovarValor)

tweets_ods.head(10)
```
