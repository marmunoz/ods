Vamos a generar una nube de palabras para utilizar en la presentación.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
tweets_limpios = pd.read_feather('../data/tweets_limpios')
tweets_limpios.head()
```

[wordcloud - Minimal Example](https://amueller.github.io/word_cloud/auto_examples/simple.html#sphx-glr-auto-examples-simple-py)

```python
text = ' '.join(tweets_limpios['text'])
```

```python
wordcloud = WordCloud().generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
wordcloud = WordCloud(max_font_size=80, width=1600, height=800, background_color='white').generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file('../img/wordcloud.png')
```

```python
x, y = np.ogrid[:300, :300]

mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

wc = WordCloud(background_color="white", mask=mask)
wc.generate(text)

plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```
